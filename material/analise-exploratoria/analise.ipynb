
**Descrição de colunas:**
- `Transaction ID:` Número identificador para cada transação.
- `Costomer ID:` Número identificador para cada cliente.
- `Transaction Amount:` Valor da compra. 
- `Transaction Date:` Data e hora da compra. 
- `Payment Method:` Método de pagamento (cartão de crédito, paypal, transferência bancária)
- `Product Category:` Categoria do produto. 
- `Quantity:` Quantia de produtos adquiridos. 
- `Costomer Age:` Idade do cliente.
- `Customer Location:` Endereço do cliente.
- `Device Used:` Dispositivo utilizado para realizar a compra (Tablet, celular, computador)
- `IP Address:` Endereço de IP do dispositivo usado para realizar a compra. 
- `Shipping Address:` Endereço de entrega da compra. 
- `Billing Address:` Endereço de cobrança ou envio da fatura do cartão. 
- `Is Fraudulent:` Indicador binário de se a transação é fraudulenta (1 para fraudulenta, 0 para legítima).
- `Account Age Days:` Há quantos dias a conta foi criada. 
- `Transaction Hour:` Hora do dia em que a compra foi feita.

**Análise exploratória sobre pedidos fraudulentos**

```python
#Mostra os dados apenas dos pedidos fraudulentos 
pedidos_fraudulentos = df_dados[df_dados['Is Fraudulent'] == 1].head(50)
pedidos_fraudulentos[['Is Fraudulent', 'Account Age Days', 'Transaction Hour', 'Transaction Date', 'Customer Age', 'Device Used', 'Product Category', 'Payment Method']]
```

```python
#Média do horário em que eram feitas as transações fraudulentas
media = pedidos_fraudulentos[['Transaction Hour']].mean()
media
```

```python
#Mostra a quantia por categoria nas colunas, nesta análise foi possível 
# visualizar que a maioria das transações foram feitas com pagamento em débito, na compra de vestuários.
quantia_categoria = pedidos_fraudulentos[['Payment Method', 'Product Category']].value_counts().reset_index(name='Count').head(10)
quantia_categoria
```

```python
#Verificando se os IP's utilizados nos pedidos fraudulentos se repetem, através dessa 
# análise é possível ver que não há recorrência de IP's
ip_fraudulentos = pedidos_fraudulentos['IP Address'].value_counts()
ip_fraudulentos
```

```python
#Ticket médio de $700.18 
ticket_medio_fraudado = pedidos_fraudulentos[pedidos_fraudulentos['Is Fraudulent'] == 1]['Transaction Amount'].mean()
ticket_medio_fraudado
```

**Análise exploratória sobre os pedidos não fraudulentos**

```python
#Retornando apenas os pedidos não fraudulentos, para fazer uma comparação e tentar identificar anomalias. 
pedidos_nao_fraudulentos = df_dados[df_dados['Is Fraudulent'] == 0].head(10)
pedidos_nao_fraudulentos[['Payment Method', 'Product Category']]
```

```python
#Nos pedidos fraudulentos, a forma de pagamento que mais retorna é com cartão de crédito, 
#enquanto nos pedidos não fraudulentos, a maioria das transações são feitas com cartão de débito ou transferências bancárias. 
nf_recorrencia_produto_pagamentos = pedidos_nao_fraudulentos[['Payment Method', 'Product Category']].value_counts()
nf_recorrencia_produto_pagamentos
```

## **Gerando gráficos**

```python

fig, ax = plt.subplots(figsize=(10, 6))

for method in quantia_categoria['Payment Method'].unique():
    subset = quantia_categoria[quantia_categoria['Payment Method'] == method]
    ax.bar(subset['Product Category'], subset['Count'], label=method)

ax.set_xlabel('Product Category')
ax.set_ylabel('Count')
ax.set_title('Quantidade de Pedidos Fraudulentos por Categoria de Produto e Método de Pagamento')

ax.legend(title='Payment Method')
plt.show()
```

Através do gráfico gerado, é possível ver as proporções entre métodos de pagamentos mais utilizados e as categorias mais visadas pelos fraudadores de acordo com a nossa base de dados, podendo concluir-se que a categoria vestuário é a mais visada, com forma de pagamento em débito. 

```python
import matplotlib.pyplot as plt

pedidos_nao_fraudulentos = df_dados[df_dados['Is Fraudulent'] == 0]

quantia_nao_fraudulentos = pedidos_nao_fraudulentos.groupby(['Payment Method', 'Product Category']).size().reset_index(name='Count')

fig, ax = plt.subplots(figsize=(10, 6))


for method in quantia_nao_fraudulentos['Payment Method'].unique():
    subset = quantia_nao_fraudulentos[quantia_nao_fraudulentos['Payment Method'] == method]
    ax.bar(subset['Product Category'], subset['Count'], label=method)

ax.set_xlabel('Product Category')
ax.set_ylabel('Count')
ax.set_title('Quantidade de Pedidos Não Fraudulentos por Categoria de Produto e Método de Pagamento')

ax.legend(title='Payment Method')
plt.show()
```

Aqui podemos ver que no geral, vestuário é a categoria mais comprada e a forma de pagamento mais utilizada em todas as categorias, em sua maioria esmagadora também é o cartão de débito. 

```python
#Proporção de valores dos pedidos fraudulentos e não fraudulentos
plt.figure(figsize=(10, 6))

plt.hist(pedidos_fraudulentos['Transaction Amount'], bins=20, alpha=0.5, color='red', label='Fraudulent Orders')

plt.hist(pedidos_nao_fraudulentos['Transaction Amount'], bins=20, alpha=0.5, color='green', label='Non-Fraudulent Orders')

plt.xlabel('Valor da Transação')
plt.ylabel('Frequência')
plt.title('Comparação da Distribuição dos Valores das Transações')
plt.legend()
plt.grid()
plt.show()
```

Esse gráfico nos mostra como os valores das compras em ambas comparações são em sua maioria de baixo valor, algo que se iguala geralmente ao cenário real, pois em grande parte, o fraudador faz compras de baixo valor. 

```python
#proporção de pedidos fraudulentos e não fraudulentos
proporcao_fraude = df_dados['Is Fraudulent'].value_counts()

proporcao_fraude.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Proporção de Pedidos Fraudulentos e Não Fraudulentos')
plt.ylabel('')
plt.show()
```

Nota-se, que no cenário da nossa base de dados, a prporção de pedidos fraudulentos ainda é menor que os pedidos não fraudulentos. 

## **Conclusão da análise**

Com 7 anos de experiência no setor, posso afirmar que o cenário de fraudes em e-commerce da nossa base de dados, embora preocupante, ainda não ultrapassou os pedidos legítimos, cenário este que se iguala ao mundo real até o momento. Contudo, é vital que analistas e profissionais da área estejam sempre atentos às inovações tecnológicas e tendências emergentes. A evolução das ferramentas de detecção de fraudes, incluindo o uso de inteligência artificial e machine learning, é essencial para manter esse equilíbrio.

Considerando as análises realizadas neste material, e focando apenas nesse cenário e nos dados disponíveis, pode-se afirmar que a maioria das transações fraudulentas pode ser classificada como autofraudes. Autofraude refere-se à situação em que o cliente efetua a compra, mas alega não tê-la feito, buscando evitar o pagamento ou solicitar o estorno do valor. Além disso, também se observa que muitos desacordos comerciais ocorrem, onde o cliente, por algum motivo, desiste da compra, opta pela troca do produto ou apresenta outras razões, e em vez de cancelar o pedido, nega a realização da compra, na expectativa de que o cancelamento aconteça de forma imediata.  

Lembrando que a conclusão das classificações dos pedidos fraudulentos foram feitas apenas sob percepções de alguém experiente, no mundo real essas classificações são feitas por uma equipe especializada que entram em contato com o cliente para saber a história verídica e juntamente com alguns insumos disponibilizados também através de dados aos quais não foram fornecidos nessa base, chegam a uma conclusão, que mesmo assim, podem não ter 100% de veracidade. 

As análises de fraude não apenas salvaguardam a integridade dos negócios, mas também geram milhões de reais em economia anualmente para grandes comércios e varejistas. Sem essas análises, as perdas poderiam ser drasticamente prejudiciais. Portanto, é de extrema import6ancia que os analistas de fraudes permaneçam um passo à frente, investindo em conhecimento e tecnologia para garantir um futuro mais seguro para o e-commerce.
