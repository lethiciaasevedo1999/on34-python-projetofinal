
**Descrição de colunas:**
- `Transaction ID:` Número identificador para cada transação.
- `Costomer ID:` Número identificador para cada cliente.
- `Transaction Amount:` Valor da compra. 
- `Transaction Date:` Data e hora da compra. 
- `Payment Method:` Método de pagamento (cartão de crédito, paypal, transferência bancária)
- `Product Category:` Categoria do produto. 
- `Quantity:` Quantia de produtos adquiridos. 
- `Costomer Age:` Idade do cliente.
- `Customer Location:` Endereço do cliente.
- `Device Used:` Dispositivo utilizado para realizar a compra (Tablet, celular, computador)
- `IP Address:` Endereço de IP do dispositivo usado para realizar a compra. 
- `Shipping Address:` Endereço de entrega da compra. 
- `Billing Address:` Endereço de cobrança ou envio da fatura do cartão. 
- `Is Fraudulent:` Indicador binário de se a transação é fraudulenta (1 para fraudulenta, 0 para legítima).
- `Account Age Days:` Há quantos dias a conta foi criada. 
- `Transaction Hour:` Hora do dia em que a compra foi feita.

## **Gerando Insights**

```python
#Contando o número de fraudes no total
numero_fraudes = df_dados['Is Fraudulent'].sum()
numero_fraudes
```

```python
#Fraudes e não fraudes
# 0 para fraude
# 1 para não foi fraude
fraude = df_dados['Is Fraudulent'].value_counts()
fraude
```

```python
#Endereços de entrega que mais retornam 
endereco = df_dados['Shipping Address'].value_counts().head(10)
endereco
```

```python
#Mostra os dados apenas dos pedidos fraudulentos 
pedidos_fraudulentos = df_dados[df_dados['Is Fraudulent'] == 1].head(50)
pedidos_fraudulentos[['Is Fraudulent', 'Account Age Days', 'Transaction Hour', 'Transaction Date', 'Customer Age', 'Device Used', 'Product Category', 'Payment Method']]
```

```python
#Média do horário em que eram feitas as transações fraudulentas
media = pedidos_fraudulentos[['Transaction Hour']].mean()
media
```

```python
#Mostra a quantia por categoria nas colunas, nesta análise foi possível 
#visualizar que a maioria das transações foram feitas com cartão de crédito, na compra de vestuários.
quantia_categoria = pedidos_fraudulentos[['Payment Method', 'Product Category']].value_counts().reset_index(name='Count')
quantia_categoria
```

```python
#Verificando se os IP's utilizados nos pedidos fraudulentos se repetem, através dessa 
# análise é possível ver que não há recorrência de IP's
ip_fraudulentos = pedidos_fraudulentos['IP Address'].value_counts()
ip_fraudulentos
```

```python
#Ticket médio de $700.18 
ticket_medio_fraudado = pedidos_fraudulentos[pedidos_fraudulentos['Is Fraudulent'] == 1]['Transaction Amount'].mean()
ticket_medio_fraudado
```

```python
#Retornando apenas os pedidos não fraudulentos, para fazer uma comparação e tentar identificar anomalias. 
pedidos_nao_fraudulentos = df_dados[df_dados['Is Fraudulent'] == 0].head(50)
pedidos_nao_fraudulentos[['Payment Method', 'Product Category']]
```

```python
#Nos pedidos fraudulentos, a forma de pagamento que mais retorna é com cartão de crédito, 
#enquanto nos pedidos não fraudulentos, a maioria das transações são feitas com cartão de débito ou transferências bancárias. 
nf_recorrencia_produtos_pagamentos = pedidos_nao_fraudulentos[['Payment Method', 'Product Category']].value_counts()
nf_recorrencia_produtos_pagamentos
```
